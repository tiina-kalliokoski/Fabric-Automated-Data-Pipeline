```python
silver_df = spark.table("silver_employee")

display(silver_df)

#sama asia voidaan tehdä näin: gold_source_df = spark.table("silver_employee")
```

```python
from pyspark.sql.functions import count, sum, avg

gold_department = (
    silver_df
    .groupBy("Department")
    .agg(
        count("*").alias("Headcount"),
        sum("AttritionFlag").alias("AttritionCount"),
        avg("AttritionFlag").alias("AttritionRate")
    )
)

display(gold_department)
```

```python
#tallennus:
gold_department.write \
    .mode("overwrite") \
    .saveAsTable("gold_department_metrics")
```

```python
gold_jobrole = (
    silver_df
    .groupBy("JobRole")
    .agg(
        count("*").alias("Headcount"),
        sum("AttritionFlag").alias("AttritionCount"),
        avg("AttritionFlag").alias("AttritionRate"),
        avg("MonthlyIncome").alias("AverageSalary")
    )
)





display(gold_jobrole)
```
