```python
# Welcome to the Silver notebook!

bronze_df = spark.table("bronze_employee")

```

```python
try:
    run_timestamp
except:
    run_timestamp = "manual_execution"

try:
    environment
except:
    environment = "DEV"

print(f"Run Timestamp: {run_timestamp}")
print(f"Environment: {environment}")
```

```python
print(bronze_df.count())
```

```python
bronze_df.printSchema()
```

```python
silver_df = bronze_df
```

```python
from pyspark.sql.functions import lit
silver_df = (
    silver_df
    .withColumn("pipeline_run_timestamp", lit(run_timestamp))
    .withColumn("environment", lit(environment))
)
```

```python
display(silver_df)
```

```python
from pyspark.sql.functions import when, col

silver_df = silver_df.withColumn(
    "AttritionFlag",
    when(col("Attrition") == "Yes", 1)
    .otherwise(0)
)
```

```python
silver_df = silver_df.withColumn(
    "SalaryBand",
    when(col("MonthlyIncome") < 5000, "Low")
    .when(col("MonthlyIncome") < 10000, "Medium")
    .otherwise("High")
)
```

```python
silver_df = silver_df.withColumn(
    "TenureGroup",
    when(col("YearsAtCompany") <= 2, "0-2 Years")
    .when(col("YearsAtCompany") <= 5, "3-5 Years")
    .when(col("YearsAtCompany") <= 10, "6-10 Years")
    .otherwise("10+ Years")
)
```

```python
silver_df = silver_df.withColumn(
    "AgeGroup",
    when(col("Age") < 30, "Under 30")
    .when(col("Age") < 40, "30-39")
    .when(col("Age") < 50, "40-49")
    .otherwise("50+")
)
```

```python
silver_df
```

```python
silver_df = silver_df.withColumn(
    "OvertimeFlag",
    when(col("OverTime") == "Yes", 1)
    .otherwise(0)
)
```

```python
display(
    silver_df.select(
        "EmployeeNumber",
        "Attrition",
        "AttritionFlag",
        "SalaryBand",
        "TenureGroup",
        "AgeGroup",
        "OvertimeFlag"
    )
)
```

```python
silver_df.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("silver_employee")
```

```python
from pyspark.sql.functions import avg

display(
    silver_df.groupBy("Department")
    .agg(
        avg("AttritionFlag")
        .alias("AttritionRate")
    )
)
```
